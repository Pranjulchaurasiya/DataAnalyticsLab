```python
import numpy as np
```

```python
# Input matrix size
rows = int(input("Enter number of rows: "))
cols = int(input("Enter number of columns: "))

print("Enter elements of Matrix A:")
A = []
for i in range(rows):
    row = list(map(int, input().split()))
    A.append(row)

print("Enter elements of Matrix B:")
B = []
for i in range(rows):
    row = list(map(int, input().split()))
    B.append(row)

A = np.array(A)
B = np.array(B)

print("\nMatrix A:\n", A)
print("Matrix B:\n", B)

```

```python
# Addition & Subtraction
if A.shape == B.shape:
    print("\nAddition:\n", A + B)
    print("\nSubtraction:\n", A - B)
else:
    print("\nAddition/Subtraction not possible")

# Multiplication
if A.shape[1] == B.shape[0]:
    print("\nMultiplication:\n", np.dot(A, B))
else:
    print("\nMultiplication not possible")

# Transpose
print("\nTranspose of A:\n", A.T)

# Inverse
if rows == cols:
    if np.linalg.det(A) != 0:
        print("\nInverse of A:\n", np.linalg.inv(A))
    else:
        print("\nInverse not possible (determinant = 0)")
else:
    print("\nInverse not possible (not a square matrix)")
```
